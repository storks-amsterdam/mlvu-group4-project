```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import rgb2hex
```

```python
# 5 min timestamps
points = [
        (52.36262230643297, 4.874122529150264),
        (52.362579158102086, 4.873932632809927),
        (52.36254163837501, 4.8737195271169504),
        (52.362523326915365, 4.874223994460076),
        (52.362130566861296, 4.873868511042837),
        (52.36215969034517, 4.873570456012154),
        (52.36268212437266, 4.875050459717635),
        (52.36303026995638, 4.876615792735298),
        (52.36261820585393, 4.873301915541297),
        (52.36320347838573, 4.876682177373811),
        ]

values = [
0.022084418684244156,
0.017296040430665016,
0.02576526626944542,
0.020556703209877014,
0.01660170964896679,
0.02055089920759201,
0.016870878636837006,
0.02422153577208519,
0.03256247192621231,
0.018515000119805336,
]


y, x = zip(*points) # creates array x with all x values and array y with all y values

epsilon = 0.0005
min_x, max_x = min(x) - epsilon, max(x) + epsilon
min_y, max_y = min(y) - epsilon, max(y) + epsilon

# Create a grid of points
x_res = np.linspace(min_x, max_x, 100) # creates an array with 100 points between 0 and 1
y_res = np.linspace(min_y, max_y, 100)

x_grid, y_grid = np.meshgrid(x_res, y_res) # creates a grid of points with x_res and y_res

# Interpolate the values using RBF
rbf = sp.interpolate.Rbf(x, y, values, function='linear') # creates a function that interpolates the values
z_grid = rbf(x_grid, y_grid) # applies the function to the grid of points


# Plot the result
plt.contourf(x_grid, y_grid, z_grid, cmap='viridis', levels=100)
cbar = plt.colorbar()
cbar.set_label('mean absolute error', rotation=90)
plt.scatter(x, y, c=values, cmap='viridis', edgecolor='k', )
plt.show()

# print(len(z_grid)) # prints the interpolated values
```
